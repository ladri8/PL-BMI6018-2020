## Inline for loop in python

```python
#Double numbers using inline for:

myList = [1,4,-1]
doubled = [num * 2 for num in myList]
print(doubled)

#The above command equals to the followling:

myList = [1,4,-1]
for num in myList:
    doubled = num * 2
    print(doubled)
```

## Inline for loop in python example

<b> Make a new function that counts only word over 5 letters long. </b>

```python
def five_words(lst):
    words = []
    for word in lst:
        if len(word) > 5:
           words.append(word)
    return words
myList=['1','111','111111','11111111']
print(five_words(myList))
```

```python
#In-line version:
def five_inlineFor_words(lst):
    return [word for word in lst if len(word) > 5]
myList=['1','111','111111','11111111']
print(five_inlineFor_words(myList))
```

## Nested List (List of Lists)

```python
L=[[1,2,-1],[3,-4,7,9],['Hi','-1',(1,2,4), {'a':1,'b':2,'d':3}]]

#Print the element in the first row and the colon
print(L[0][1])

#Print colons and rows
for ele in L:
    print(*ele)
```

## Shallow and deep copies (examples in List)




<b> It could be for any sequence type.

```python
#Shallow copy

L1=[2,4,5]
L2=L1
print(L1,L2)
L1[1]=3
print(L1,L2)
```

```python
#Deep copy

import copy

help(copy)
```

```python
L1=[2,4,5]
L2=copy.deepcopy(L1)

print(L1,L2)
L1[1]=3
print(L1,L2)
```

# Replace key with value and value with key:

```python
myDict={1:'a',2:'b'}
new_dict = {value:key for (key,value) in myDict.items()}
print(new_dict)
```
