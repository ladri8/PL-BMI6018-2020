## Question 2: Abstract class using ABC and   abstractmethod method (30 points)

In the below example (taken from the lecture 11-04-2020):
1. Implement an abstract class inherited from ABC. Call it Country.  (5 points)
2. In the Country class, define def \__init\__(self, value) to enter population attribute. (5 points)
3. In the Country class, define abstract methods of capital, language, and type. (5 points)
4. Try to instantiate abstract class Country to see if it raises an error. (1 point)
5. Run the code without step 4 (comment step 4). (1 point)
6. Determine the piece of code that shows polymorphism (6 points) and the piece of code that shows method overriding (7 points). 


```python
#Write an abstract class with its abstract methods

class India(Country): 
    def capital(self): 
        print("New Delhi is the capital of India with a population  "+ str(self._population)) 
    def language(self): 
        print("Hindi is the most widely spoken language of India.") 
   
    def type(self): 
        print("India is a developing country.") 

class USA(Country): 
    def capital(self): 
        print("Washington, D.C. is the capital of USA with a population  "+ str(self._population)) 
   
    def language(self): 
        print("English is the primary language of USA.") 
   
    def type(self): 
        print("USA is a developed country.") 

def func(obj): 
    obj.capital() 
    obj.language() 
    obj.type() 


#Try to instantiate abstract class Country to see if it raises an error.
obj_ind = India(1353000000) 
obj_usa = USA(328200000) 


func(obj_ind) 
func(obj_usa) 

#Run the code without step 4 (comment step 4)
#Determine the piece of code that shows polymorphism  and the piece of code that shows method overriding. 
```
